```python
# Pythonでデータ可視化（Matplotlib）を勉強するならまずこの動画で特訓
# https://www.youtube.com/watch?v=bsYJ3hTvx7c

import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

# 日本語フォント設定
from matplotlib import rc
jp_font = "Yu Gothic"
rc('font', family=jp_font)

import numpy as np
```

```python
# 01. 折れ線グラフ

x = np.arange(10)
y = np.random.randint(-10, 10, 10)

# xが無い場合は、ｙのインデックスに従う
plt.plot(y)  
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(y)  
```

```python
# 02. 散布図

x1 = np.random.randint(10, 20, 20)
x2 = np.random.randint(20, 30, 20)
y1 = np.random.randint(50, 100, 20)
y2 = np.random.randint(0, 40, 20)

plt.scatter(x1, y1)
plt.scatter(x2, y2)
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.scatter(x1, y1)
ax.scatter(x2, y2)
```

```python
# 03. ヒストグラム

data = np.random.randn(100)

plt.hist(data, bins=10, range=(-5, 5))
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.hist(data, bins=10, range=(-5, 5))
```

```python
# 04. 棒グラフ

x = ["Sam", "John", "Kevin", "Adam"]
y = np.random.randint(0, 200, 4)

plt.bar(x, y)
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.bar(x, y)
```

```python
# 05. タイトルとラベル

x = np.arange(10)
y = np.random.randint(-10, 10, 10)

plt.plot(x, y)

plt.title("結果")
plt.xlabel("X軸")
plt.ylabel("Y軸")
plt.grid(True)
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x, y)
ax.set_title("結果")
ax.set_xlabel("X軸")
ax.set_ylabel("Y軸")
ax.grid(True)
```

```python
# 06. 軸範囲

x1 = np.random.randint(10, 20, 20)
x2 = np.random.randint(20, 30, 20)
y1 = np.random.randint(50, 100, 20)
y2 = np.random.randint(0, 40, 20)

plt.scatter(x1, y1)
plt.scatter(x2, y2)
plt.xlim(0, 40)
plt.ylim(0, 120)
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.scatter(x1, y1)
ax.scatter(x2, y2)
ax.set_xlim(0, 40)
ax.set_ylim(0, 120)
```

```python
# 07. 対数軸

x = np.linspace(0, 10, 500)
y = np.exp(x)

plt.plot(x, y)
plt.yscale("log")
plt.grid(which="major")
plt.grid(which="minor")
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x, y)
ax.set_yscale("log")
ax.grid(which="major")
ax.grid(which="minor")
```

```python
# 08. 凡例

x = np.linspace(0, 2*np.pi, 500)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin")
plt.plot(x, y2, label="sin")
plt.ylim(-2, 2)
plt.legend(loc=3)  # loc=0:best 1:upper right 2:upper left 3:lower lreft
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x, y1, label="sin")
ax.plot(x, y2, label="sin")
ax.set_ylim(-2, 2)
ax.legend(loc=3)
```

```python
# 09. マーカの種類・サイズ変更

x1 = np.random.randint(10, 35, 20)
x2 = np.random.randint(5, 45, 20)
x3 = np.random.randint(0, 40, 20)
y1 = np.random.randint(50, 100, 20)
y2 = np.random.randint(0, 40, 20)
y3 = np.random.randint(20, 80, 20)

plt.scatter(x1, y1, s=80, marker="*")
plt.scatter(x2, y2, s=60, marker="^")
plt.scatter(x3, y3, s=30, marker="x")
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.scatter(x1, y1, s=80, marker="*")
ax.scatter(x2, y2, s=60, marker="^")
ax.scatter(x3, y3, s=30, marker="x")
```

```python
# 10. 注釈

data = [5, 3, 4, 2, 0, 3, 2, 1, 4, 6, 8, 5]

plt.plot(data)
plt.annotate("min value", xy=(4.2, 0), xytext=(9, 1), arrowprops=dict(facecolor="black", shrink=0.1))
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(data)
ax.annotate("min value", xy=(4.2, 0), xytext=(9, 1), arrowprops=dict(facecolor="black", shrink=0.1))
```

```python
# 11. 目盛り

data = [5, 3, 4, 2, 0, 3, 2, 1, 4, 6, 8, 5]

plt.plot(data)
plt.annotate("min value", xy=(4.2, 0), xytext=(9, 1), arrowprops=dict(facecolor="black", shrink=0.1))
_ = plt.xticks(np.arange(12))
_ = plt.yticks(np.arange(0, 10, 2))
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(data)
ax.annotate("min value", xy=(4.2, 0), xytext=(9, 1), arrowprops=dict(facecolor="black", shrink=0.1))
_ = ax.set_xticks(np.arange(12))
_ = ax.set_yticks(np.arange(0, 10, 2))
```

```python
# 13. 複数グラフの表示

names = ["Sam", "John", "Kevin", "Adam"]
values = [60, 170, 10, 120]

plt.figure(figsize=(9, 3))
plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
```

```python
fig = plt.figure(figsize=(9,3))
ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)

ax1.bar(names, values)
ax2.scatter(names, values)
ax3.plot(names, values)
```

```python
fig, ax = plt.subplots(1,3, figsize=(9,3))
 
ax[0].bar(names, values)
ax[1].scatter(names, values)
ax[2].plot(names, values)
```
