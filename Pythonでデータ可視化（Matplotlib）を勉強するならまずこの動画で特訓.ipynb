```python
# 08. 凡例

x = np.linspace(0, 2*np.pi, 500)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin")
plt.plot(x, y2, label="sin")
plt.ylim(-2, 2)
plt.legend(loc=3)  # loc=0:best 1:upper right 2:upper left 3:lower lreft
```

```python
# 15. 3次元グラフ

from mpl_toolkits.mplot3d import Axes3D

theta = np.linspace(-4*np.pi, 4*np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)


fig = plt.figure()
ax = fig.gca(projection="3d")

ax.plot(x, y, z, label="3D Curve")
ax.legend()
```

```python
fig = plt.figure()
ax = Axes3D(fig)

ax.plot(x, y, z, marker="o", linestyle="None", label="3D Curve")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.legend()
```

```python
# 18. 演習問題1

x1 = np.arange(10)
y1 = np.random.randint(-10, 10, 10)
x2 = ["Sam", "John", "Kevin", "Adam"]
y2 = np.random.randint(0, 200, 4)

plt.figure(figsize=(9, 6))

plt.subplot(211)
plt.plot(x1, y1)
plt.xlabel("x軸")
plt.ylabel("y軸")

plt.subplot(212)
plt.bar(x2, y2)
```

```python
fig = plt.figure(figsize=(9,6))
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

ax1.plot(x1, y1)
ax1.set_xlabel("x軸")
ax1.set_ylabel("y軸")

ax2.bar(x2, y2)
```

```python
fig, ax = plt.subplots(2,1, figsize=(9,6))
 
ax[0].plot(x1, y1)
ax[0].set_xlabel("x軸")
ax[0].set_ylabel("y軸")

ax[1].bar(x2, y2)
```

```python
# 19. 演習問題2

labels = ["Frogs", "Hogs", "Dogs", "Logs"]
sizes = [15, 30, 45, 10]
explode = [0, 0.1, 0, 0]

plt.pie(sizes, labels=labels, counterclock=True, startangle=90, explode=explode, shadow=True, autopct="%1.1f%%")
_ = plt.axis("equal")
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.pie(sizes, labels=labels, counterclock=True, startangle=90, explode=explode, shadow=True, autopct="%1.1f%%")
_ = ax.axis("equal")
```

```python
# 20. 演習問題3

mu1, sigma1 = 100,15
mu2, sigma2 = 90, 20
mu3, sigma3 = 110, 10
x1 = mu1 + sigma1 * np.random.randn(100)
x2 = mu2 + sigma2 * np.random.randn(100)
x3 = mu3 + sigma3 * np.random.randn(100)

labels = ["x1", "x2", "x3"]

plt.hist([x1, x2, x3], label=labels, histtype="barstacked", density=True, bins=10, range=(0, 180))
plt.title("ヒストグラム")
plt.xlabel("x")
plt.ylabel("頻度")
plt.grid(True)
_ = plt.legend(loc=2)
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.hist([x1, x2, x3], label=labels, histtype="barstacked", density=True, bins=10, range=(0, 180))
ax.set_title("ヒストグラム")
ax.set_xlabel("x")
ax.set_ylabel("頻度")
ax.grid(True)
_ = ax.legend(loc=2)
```
