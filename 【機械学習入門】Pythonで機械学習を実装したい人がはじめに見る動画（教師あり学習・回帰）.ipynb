```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 
from matplotlib import rc
jp_font = "Yu Gothic"
rc('font', family=jp_font)

import seaborn as sns
```

```python
# 

from sklearn.datasets import load_boston
dataset = load_boston()

dataset
```

```python
# 

x, t = dataset.data, dataset.target
columns = dataset.feature_names

print(type(x), np.shape(x), "\n")
print(x, "\n")
print(type(t), np.shape(t),  "\n")
print(t[:20], "\n")                 # 20
print(columns)
```

```python
# CRIM	
# ZN	25,000
# INDUS	
# CHAS	
# NOX	
# RM	
# AGE	1940
# DIS	5
# RAD	
# TAX	10,000
# PTRATIO	
# B	
# LSTAT	
# MEDV	1,000
```

```python
# numpy ArrayPandas

df = pd.DataFrame(x, columns=columns)

df
```

```python
# tPandas

df["Target"] = t

df
```

```python
# scikit-learnnumpy Array
# numpy Array
```

```python
# Target

t = df["Target"].values

print(type(t), np.shape(t), "\n")
print(t[:20], "\n")
```

```python
# Target
# Targetdrop

x = df.drop(labels=["Target"], axis=1).values

x
```

```python
# 
# 
```

```python
# 

from sklearn.model_selection import train_test_split
```

```python
# 7030

x_train, x_test, t_train, t_test = train_test_split(x, t, test_size=0.3, random_state=0)
# random_state
# 0

print(f"x_train:{np.shape(x_train)}  x_test:{np.shape(x_test)}")
print(f"t_train:{np.shape(t_train)}  t_test:{np.shape(t_test)}")
```

```python
# scikit-learnSTEP1
# 
```

```python
# 
# fit
# predict
# score

from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
# fit

model.fit(x_train, t_train)
```

```python
# :w

model.coef_
```

```python
plt.figure(figsize=(10, 3))
plt.bar(x=columns, height=model.coef_)
_ = plt.title(" w")
```

```python
# :b

model.intercept_
```

```python
# scikit-learnSTEP2
# 
# 20 ---- 1
```

```python
print(f"train score:{model.score(x_train, t_train)}  :")
print(f"test  score:{model.score(x_test, t_test)}  ")
```

```python
# scikit-learnSTEP3
# 
```

```python
# 

y = model.predict(x_test)
```

```python
for i in range(5):
    print(f"{y[i]:.1f}  {t_test[i]:.1f}")
```

```python
df = pd.read_csv("regression_pls.csv")
df
```

```python
# scikit-learnnumpy Array
# numpy Array

# Target
t = df["Target"].values

# Target
# Targetdrop
x= df.drop("Target", axis=1).values

print(f"t:{np.shape(t)}  x:{np.shape(x)}")
```

```python
# 
# 
# 

from sklearn.model_selection import train_test_split
```

```python
# 7030

x_train, x_test, t_train, t_test = train_test_split(x, t, test_size=0.3, random_state=0)
# random_state
# 0

print(f"x_train:{np.shape(x_train)}  x_test:{np.shape(x_test)}")
print(f"t_train:{np.shape(t_train)}  t_test:{np.shape(t_test)}")
```

```python
# scikit-learnSTEP1
# 
# 
# fit
# predict
# score

from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
# fit

model.fit(x_train, t_train)
```

```python
# scikit-learnSTEP2
# 
# 20 ---- 1

print(f"train score:{model.score(x_train, t_train)}  :")
print(f"test  score:{model.score(x_test, t_test)}  ")
```

```python

#  (Partial Least Squares, PLS) 
```

```python
from sklearn.cross_decomposition import PLSRegression

# 
pls = PLSRegression(n_components=7)
```

```python
# PLSfit

pls.fit(x_train, t_train)
```

```python
print(f"train score:{pls.score(x_train, t_train)}  :")
print(f"test  score:{pls.score(x_test, t_test)}  ")
```

```python
# 

for i in range(5, 15, 1):
    pls = PLSRegression(n_components=i)
    pls.fit(x_train, t_train)
    print(f"n_components:{i}  train_score:{pls.score(x_train, t_train):0.2f}  test_score:{pls.score(x_test, t_test):0.2f}")
```
