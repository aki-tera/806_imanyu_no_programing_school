```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 日本語フォント設定
from matplotlib import rc
jp_font = "Yu Gothic"
rc('font', family=jp_font)

import seaborn as sns
```

```python
# ボストンの住宅データを取得

from sklearn.datasets import load_boston
dataset = load_boston()

dataset
```

```python
# データの取り出し

x, t = dataset.data, dataset.target
columns = dataset.feature_names

print(type(x), np.shape(x), "\n")
print(x, "\n")
print(type(t), np.shape(t),  "\n")
print(t[:20], "\n")                 # 先頭20個のみ表示
print(columns)
```

```python
# CRIM	犯罪発生率
# ZN	25,000平方フィート以上の住宅区画の割合
# INDUS	非小売業種の土地面積の割合
# CHAS	チャールズ川沿いかを表すダミー変数
# NOX	窒素酸化物の濃度
# RM	平均部屋数
# AGE	1940年より前に建てられた建物の割合
# DIS	5つのボストンの雇用施設への重み付き距離
# RAD	高速道路へのアクセスのしやすさ
# TAX	10,000ドルあたりの不動産税率
# PTRATIO	生徒と教師の割合
# B	黒人の割合
# LSTAT	低所得者の割合
# MEDV	住宅価格の中央値（1,000単位）
```

```python
# データがnumpy Arrayなので、Pandasに変換する

df = pd.DataFrame(x, columns=columns)

df
```

```python
# tをPandasに追加

df["Target"] = t

df
```

```python
# scikit-learnは基本的にnumpy Arrayで処理する
# そのため、numpy Arrayに変換すること
```

```python
# Targetを出力値として設定する

t = df["Target"].values

print(type(t), np.shape(t), "\n")
print(t[:20], "\n")
```

```python
# Target以外を入力値として設定する
# そのため、Targetをdropで除外

x = df.drop(labels=["Target"], axis=1).values

x
```

```python
# 準備
# 学習データを検証データを準備する
```

```python
# 学習するデータと検証するデータを分ける必要がある

from sklearn.model_selection import train_test_split
```

```python
# 学習データの割合を70％、検証データの割合を30％とする

x_train, x_test, t_train, t_test = train_test_split(x, t, test_size=0.3, random_state=0)
# random_stateは取り出すデータをランダムに出すシードの値
# 0にすることで毎回同じになる

print(f"x_train:{np.shape(x_train)}  x_test:{np.shape(x_test)}")
print(f"t_train:{np.shape(t_train)}  t_test:{np.shape(t_test)}")
```

```python
# 【scikit-learn】STEP1
# モデルの定義を行う
```

```python
# 線形回帰モデル
# fitメソッド：線形モデルの重みを学習
# predictメソッド：線形モデルから目的変数を予測
# scoreメソッド：決定係数（線形モデルがどの程度目的変数を説明できるか）を出力

from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
# fitメソッドで重みを学習する

model.fit(x_train, t_train)
```

```python
# 説明変数の係数:w

model.coef_
```

```python
plt.figure(figsize=(10, 3))
plt.bar(x=columns, height=model.coef_)
_ = plt.title("説明変数の係数 w")
```

```python
# 線形モデルの切片:b

model.intercept_
```

```python
# 【scikit-learn】STEP2
# 今回の学習済みデータの検証を行う
# 決定係数（相関係数の2乗）は、『0が悪い ＜----＞ 1が良い』
```

```python
print(f"train score:{model.score(x_train, t_train)}  ←学習に使用した結果:学習データが悪いので少し過学習がおきている")
print(f"test  score:{model.score(x_test, t_test)}  ←未学習データの結果")
```

```python
# 【scikit-learn】STEP3
# 推論した結果を確認する
```

```python
# 学習した結果を使って予測値を確認する

y = model.predict(x_test)
```

```python
for i in range(5):
    print(f"予測値：{y[i]:.1f}  ←　実測値：{t_test[i]:.1f}")
```

```python
df = pd.read_csv("regression_pls.csv")
df
```

```python
# scikit-learnは基本的にnumpy Arrayで処理する
# そのため、numpy Arrayに変換すること

# Targetを出力値として設定する
t = df["Target"].values

# Target以外を入力値として設定する
# そのため、Targetをdropで除外
x= df.drop("Target", axis=1).values

print(f"t:{np.shape(t)}  x:{np.shape(x)}")
```

```python
# 準備
# 学習データを検証データを準備する
# 学習するデータと検証するデータを分ける必要がある

from sklearn.model_selection import train_test_split
```

```python
# 学習データの割合を70％、検証データの割合を30％とする

x_train, x_test, t_train, t_test = train_test_split(x, t, test_size=0.3, random_state=0)
# random_stateは取り出すデータをランダムに出すシードの値
# 0にすることで毎回同じになる

print(f"x_train:{np.shape(x_train)}  x_test:{np.shape(x_test)}")
print(f"t_train:{np.shape(t_train)}  t_test:{np.shape(t_test)}")
```

```python
# 【scikit-learn】STEP1
# モデルの定義を行う
# 線形回帰モデル
# fitメソッド：線形モデルの重みを学習
# predictメソッド：線形モデルから目的変数を予測
# scoreメソッド：決定係数（線形モデルがどの程度目的変数を説明できるか）を出力

from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
# fitメソッドで重みを学習する

model.fit(x_train, t_train)
```

```python
# 【scikit-learn】STEP2
# 今回の学習済みデータの検証を行う
# 決定係数（相関係数の2乗）は、『0が悪い ＜----＞ 1が良い』

print(f"train score:{model.score(x_train, t_train)}  ←学習に使用した結果:学習データが悪いので過度な過学習がおきている")
print(f"test  score:{model.score(x_test, t_test)}  ←未学習データの結果")
```

```python

# 多重共線性（マルチコ）の対策
```

```python
# 相関関係の確認

df_corr = df.corr()
```

```python
# ヒートマップで相関関係を確認する（今回は先頭の20列を見る）

plt.figure(figsize=(12,8))
sns.heatmap(df_corr.iloc[:20, :20], annot=True)
```

```python

# 部分的最小二乗法 (Partial Least Squares, PLS) の説明
```

```python
from sklearn.cross_decomposition import PLSRegression

# まとめて少なくしたいパラメータの数
pls = PLSRegression(n_components=7)
```

```python
# PLSでまとめた結果をもとに、fitメソッドで重みを学習する

pls.fit(x_train, t_train)
```

```python
print(f"train score:{pls.score(x_train, t_train)}  ←学習に使用した結果:学習データが少し改善")
print(f"test  score:{pls.score(x_test, t_test)}  ←未学習データの結果")
```

```python
# いろいろなパラメータにした時の結果

for i in range(5, 15, 1):
    pls = PLSRegression(n_components=i)
    pls.fit(x_train, t_train)
    print(f"n_components:{i}  train_score:{pls.score(x_train, t_train):0.2f}  test_score:{pls.score(x_test, t_test):0.2f}")
```
