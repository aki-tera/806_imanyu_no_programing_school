```python
# 18問目：折れ線グラフの表示

import pandas as pd
import re
import matplotlib.pyplot as plt

# 日本語フォント設定
from matplotlib import rc
jp_font = "Yu Gothic"
rc('font', family=jp_font)


file = "weather.csv"
df =  pd.read_csv(file, encoding="utf-8")

for i in df.columns:
    if i[-2:] == ".1" or i[-2:] == ".2":
        df = df.drop(columns=i)
df = df.drop(index=0)

for i in df.columns:
    temp = re.sub("\(.+\)", "", i)
    print(temp)
    df = df.rename(columns={i:temp}) # 特定の列名のみ変更

df
```

```python
df.iloc[0:50, 0:4].plot(x=df.columns[0])
```

```python
df[:50].plot(x="年月日", y=["平均気温", "最高気温", "最低気温"], legend=False)
```
